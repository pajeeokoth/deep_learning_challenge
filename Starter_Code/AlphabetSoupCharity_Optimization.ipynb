## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
df = application_df.copy()
df = df.drop(columns=['EIN', 'NAME'], axis=1)
df.head()
```

```python
# Determine the number of unique values in each column.
df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = ['T2','T6','T5','T7','T8','T9','T10','T13','T12','T25','T14','T29','T15','T17']

# Replace in dataframe
for app in application_types_to_replace:
    df['APPLICATION_TYPE'] = df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
list = ['C2190', 'C4200','C2600', 'C5200', 'C1370', 'C1248', 'C6100', 'C1820', 'C1900', 'C1236', 'C3700', 'C2570', 'C1580', 'C1245',
                 'C2500', 'C1570', 'C1283', 'C2380', 'C1732', 'C1728', 'C2170', 'C4120', 'C8210', 'C2561', 'C4500', 'C2150']
df.drop(df[df.CLASSIFICATION.isin(list)].index, inplace=True)
df['CLASSIFICATION'].value_counts()
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = ['C7000','C1700','C4000','C5000','C1270','C2700','C2800','C7100','C1300','C1280','C1230','C1400','C7200','C2300','C1240',
                             'C8000','C7120','C1500','C1800','C6000','C1250','C8200','C1238','C1278','C1237','C1235','C7210','C4100','C1720','C2400',
                             'C1600','C1257','C1260','C2710','C0','C1267','C1256','C1234','C1246','C3200','C3000','C2100']

# Replace in dataframe
for cls in classifications_to_replace:
    df['CLASSIFICATION'] = df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
df['CLASSIFICATION'].value_counts()
```

```python
# Choose a cutoff value and create a list of organizations to be replaced
# use the variable name `incomes_to_replace`
org_to_replace = ['Association','Co-operative','Corporation']

# Replace in dataframe
for organ in org_to_replace:
    df['ORGANIZATION'] = df['ORGANIZATION'].replace(organ,"others")

# Check to make sure binning was successful
df['ORGANIZATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummy = ['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION',
         'ORGANIZATION', 'SPECIAL_CONSIDERATIONS','USE_CASE','INCOME_AMT']
df_dummies = pd.get_dummies(df, columns=dummy)

df_dummies.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = df_dummies['IS_SUCCESSFUL'].values
X = df_dummies.drop(['IS_SUCCESSFUL'],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  50
hidden_nodes_layer2 = 30
hidden_nodes_layer3 = 5

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
